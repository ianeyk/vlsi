```python
from numpy import ceil

nrows = 4
ncols = 4

control_block = ""
indent = "    "
for code in range(nrows * ncols + 1):
    
    control_statement = ""
    for row in range(nrows):
        control_statement += f"""{indent}  let y{row} = {int((code - 1) % nrows >= row)} * 1.8
"""
    for col in range(ncols):
        control_statement += f"""{indent}  let x{col} = {int(ceil(code / ncols) > col)} * 1.8
"""

    control_block += f"""
{indent}if code eq {code}
{control_statement}{indent}end"""

alter_statement = ""
for col in range(ncols):
    alter_statement += f"""{indent}alter V{col} $&x{col}
"""
for row in range(nrows):
        alter_statement += f"""{indent}alter V{ncols + row} $&y{row}
"""
    
print(f"""{control_block}

{alter_statement}""")
```
