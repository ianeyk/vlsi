```python
from numpy import ceil

nrows = 4
ncols = 4

control_block = ""
indent = "    "
for code in range(nrows * ncols + 1):
    
    control_statement = ""
    for row in range(nrows):
        control_statement += f"""{indent}  y{row} = {code % nrows > row}
"""
    for col in range(ncols):
        control_statement += f"""{indent}  x{col} = {ceil(code / ncols) > col}
"""

    control_block += f"""
{indent}if code == {code}
{control_statement}
{indent}end"""
    
print(control_block)
```
